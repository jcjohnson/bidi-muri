```python
import os, json
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
with open('data/checkpoint.t7.json', 'r') as f:
  checkpoint = json.load(f)
  
losses = np.asarray(checkpoint['train_loss_history'])
plt.plot(losses, 'o')
plt.show()
```

```python
recall = sorted([(int(k), v['recall_at_1']) for k, v in checkpoint['trainval_stat_history'].iteritems()])
ts = np.asarray([x[0] for x in recall])
ys = np.asarray([x[1] for x in recall])
plt.plot(ts, ys)
```
